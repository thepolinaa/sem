```python
# 7
def convert_input(input_str):
    try:
        num = float(input_str)
        if num.is_integer():
            return int(num)
        else:
            return int(num) + 1
    except ValueError:
        if input_str.lower() == "true":
            return True
        elif input_str.lower() == "false":
            return False
        else:
            return input_str
def fill_list(length):
    user_list = []
    for i in range(length):
        user_input = input(f"введите элемент {i + 1}: ")
        user_list.append(convert_input(user_input))
    return user_list
length = int(input("введите длину списка: "))
result_list = fill_list(length)
print(result_list)
```

```python
# 8
def merge_strings(str1, str2):
    words1 = str1.split()
    words2 = str2.split()
    merged_words = []
    for w1, w2 in zip(words1, words2):
        merged_words.append(w1)
        merged_words.append(w2)
    if len(words1) > len(words2):
        merged_words.extend(words1[len(words2):])
    elif len(words2) > len(words1):
        merged_words.extend(words2[len(words1):])
    return ' '.join(merged_words)
str1 = 'один два три'
str2 = 'альфа бетта гамма'
result = merge_strings(str1, str2)
print(result)
```

```python
# 9
def merge_strings_with_zip(str1, str2):
    words1 = str1.split()
    words2 = str2.split()
    merged_words = [word for pair in zip(words1, words2) for word in pair]
    if len(words1) > len(words2):
        merged_words.extend(words1[len(words2):])
    elif len(words2) > len(words1):
        merged_words.extend(words2[len(words1):])
    return ' '.join(merged_words)
str1 = 'один два три'
str2 = 'альфа бетта гамма'
result = merge_strings_with_zip(str1, str2)
print(result)
```

```python
# 10
def cyclic_shift_words(sentence, shift):
    words = sentence.split()
    shift = shift % len(words)
    shifted_words = words[-shift:] + words[:-shift]
    return ' '.join(shifted_words)
sentence = "один два три четыре пять шесть семь"
shift = 3
result = cyclic_shift_words(sentence, shift)
print(result)
```

```python
# 11
def prime_factors(n):
    factors = []
    i = 2
    while i*i <= n:
        if n % i:
            i += 1
        else:
            n //= i
            factors.append(i)
    if n > 1:
        factors.append(n)
    return factors
def replace_non_primes_with_factors(lst):
    result = []
    for num in lst:
        if num < 2:
            result.append(num)
        else:
            factors = prime_factors(num)
            result.extend(factors)
    return result
numbers = [3, 6, 7, 12]
result = replace_non_primes_with_factors(numbers)
print(result)
```

```python
# 12
def find_unique_min_value(lst):
    min_val = min(lst)
    count = lst.count(min_val)
    if count == 1:
        return min_val
    else:
        return None
numbers = [3, 2, 5, 2, 7, 2]
unique_min_value = find_unique_min_value(numbers)
print(unique_min_value)
```

```python
# 15
def matrix_multiplication(matrix1, matrix2):
    result = [[0 for _ in range(len(matrix2[0]))] for _ in range(len(matrix1))]
    for i in range(len(matrix1)):
        for j in range(len(matrix2[0])):
            for k in range(len(matrix2)):
                result[i][j] += matrix1[i][k] * matrix2[k][j]   
    return result
matrix1 = [[1, 2], [3, 4]]
matrix2 = [[1, 0], [0, 1]]
result_matrix = matrix_multiplication(matrix1, matrix2)
print(result_matrix)
```

```python
# 17
def sort_words_in_string(input_str):
    words = input_str.split()
    sorted_words = sorted(words)
    return ' '.join(sorted_words)
input_string = 'дыня абрикос яблоко арбуз'
sorted_string = sort_words_in_string(input_string)
print(sorted_string)
```

```python
# 18
def merge_lists(lsls):
    result_list = []
    remaining_elements = []
    i = 0
    while True:
        added = False
        for lst in lsls:
            if i < len(lst):
                result_list.append(lst[i])
                added = True
            else:
                remaining_elements.extend(lst[i:])
        if not added:
            break
        i += 1
    return result_list, remaining_elements
lsls = [[1, 2, 3], ['a', 'b'], [30, 40, 50, 60]]
result, remaining = merge_lists(lsls)
print(result)
print(remaining)
```
